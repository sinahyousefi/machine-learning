```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [3,4,2,7,9,11,2,7,8]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color = "green")
plt.title("example")
plt.xlabel("x-axis")
plt.ylabel("output")
plt.show()
```

```python
plt.plot(x,y,'o', color = "green")
plt.title("example")
plt.xlabel("x-axis")
plt.ylabel("output")
plt.show()
```

```python
plt.plot(x,y,marker='*', color = "green")
plt.title("example")
plt.xlabel("x-axis")
plt.ylabel("output")
plt.show()
```

```python
x = ["red",'green','blue','orange']
y = [5,19,7,2]
plt.bar(x,y, color = "red")
plt.title("bar-example")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
x = ["red",'green','blue','orange']
y = [5,19,7,2]
plt.pie(y, labels=x, autopct="%1.1f")
plt.title("pie-example")
plt.show()
```

```python
x = ["red",'green','blue','orange']
y = [5,19,7,2]
plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red",'green','blue','orange'])
plt.title("pie-example")
plt.show()
```

```python
import numpy as np

x = np.random.rand(1000,1)

plt.hist(x,bins=5)
plt.show()
```
